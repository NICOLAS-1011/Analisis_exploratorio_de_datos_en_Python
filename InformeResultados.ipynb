```python
import pandas as pd
```

```python
df = pd.read_csv('Data_Caso_Propuesto.csv')
```

```python
rangos = [50000000, 100000000, 150000000, 200000000, 250000000, 300000000]
nombreRango = ["A", "B", "C", "D", "E"]
df['Rango_Precios'] = pd.cut(df['Precio'], rangos,labels = nombreRango)

rangosCat = ['COMERCIAL', 'RURAL', 'INDUSTRIAL', 'UNO', 'DOS', 'TRES', 'CUATRO', 'CINCO', 'SEIS']
nombreRangoCat = ["Comerc", "rur", "indus", "Uno", "Dos", "Tres", "Cuatro", "Cinco", "Seis"]

categorias = pd.Categorical(df['Estrato'], categories=rangosCat)
df['Rango_Categorias'] = categorias.map(dict(zip(rangosCat, nombreRangoCat)))
```

```python
import matplotlib.pyplot as plt

df.plot(x='Estrato', y='Precio', kind='bar', color='blue', legend=False)
plt.xlabel('Estrato')
plt.ylabel('Precio')
plt.title('Comparacin entre estrato y precio')

plt.twinx()  # Crea un eje secundario para la lnea
df.plot(x='Estrato', y='Precio', kind='line', marker='o', color='red', ax=plt.gca(), legend=False)
plt.ylabel('Valores (lnea)', color='red')

plt.show()
```

```python
en_puja_df = df[df['Detalle Disponibilidad'] == 'EN PUJA']


count_by_tipo = en_puja_df['Tipo de Inmueble'].value_counts()

plt.figure(figsize=(8, 6))
count_by_tipo.plot(kind='bar')
plt.title('Tipos de Inmuebles en estado "En puja"')
plt.xlabel('Tipo de Inmueble')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```
